```python
# Task 01: Zigzag Walk

def walk_zigzag(floor):
    m, n = floor.shape
    for col in range(n):
        if col % 2 == 0:  # Even index columns (0-based) - top to bottom
            for row in range(m):
                if (row + col) % 2 == 0:  # Black tiles
                    print(floor[row][col], end=' ')
            print()
        else:  # Odd index columns - bottom to top
            for row in range(m - 1, -1, -1):
                if (row + col) % 2 == 0:  # Black tiles
                    print(floor[row][col], end=' ')
            print()

#DO NOT CHANGE THE CODE BELOW
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

```python
#Task 04: Matrix Compression

def compress_matrix(mat):
    r,c = mat.shape
    arr = np.zeros(((r//2),(c//2)), dtype=int)

    for i in range(0,r,2):
        for j in range(0,c,2):
            sum=mat[i][j]+mat[i+1][j]+mat[i][(j)+1]+mat[(i)+1][(j)+1]
            arr[i//2][j//2]=sum
    return arr


#DO NOT CHANGE THE CODE BELOW
matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print("...............")
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print

#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

```python
#Task 05: Game Arena

def play_game(arena):
    rows = len(arena)
    cols = len(arena[0]) if rows > 0 else 0

    total_points = 0

    for i in range(rows):
        for j in range(cols):
            if arena[i][j] % 50 == 0 and arena[i][j] != 0:
                # Check above
                if i > 0 and arena[i - 1][j] == 2:
                    total_points += 2
                # Check below
                if i < rows - 1 and arena[i + 1][j] == 2:
                    total_points += 2
                # Check left
                if j > 0 and arena[i][j - 1] == 2:
                    total_points += 2
                # Check right
                if j < cols - 1 and arena[i][j + 1] == 2:
                    total_points += 2

    if total_points >= 10:
        print(f"Points Gained: {total_points}. Your team has survived the game.")
    else:
        print(f"Points Gained: {total_points}. Your team is out.")

#DO NOT CHANGE THE CODE BELOW
arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.
print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```
