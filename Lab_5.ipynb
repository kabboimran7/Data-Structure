```python
# Task - 01:

def task1A():
    i = 1
    while i < 11:
      print(i, end=' ')
      i += 1

def task1B_recursive():
    def helper(i):
      if i < 11:
        print(i,end=' ')
        i += 1
        helper(i)
    helper(1)

def task1C():
    N = int(input('Enter a number: '))
    i = 1
    while i < N+1:
      print(i ,end=' ')
      i += 1

def task1D_recursive():
    N = int(input('Enter a number: '))
    def helper(i):
      if i < N+1:
        print(i,end=' ')
        i += 1
        helper(i)
    helper(1)
```

```python
# Task - 02:

def task2A( arr ):
  i = 0
  while i < len(arr):
    print(arr[i],end=' ')
    i+=1

def task2B_recursive( arr, i=0 ):
  if i < len(arr) :
    print(arr[i],end=' ')
    task2B_recursive( arr, i+1 )

def task2C( arr ):
    sum = 0
    i = 0
    while i < len(arr):
      sum += arr[i]
      i += 1
    return sum

def task2D_recursive( arr ,i=0, sum=0 ):
  if i < len(arr):
    sum += arr[i]
    return  task2D_recursive(arr, i+1, sum)
  return sum

def task2E( arr ):
    even_sum = 0
    odd_mul = 1
    i = 0
    while i < len(arr):
      if arr[i]%2 == 0 :
        even_sum += arr[i]
      else :
        odd_mul *= arr[i]
      i += 1
    return odd_mul - even_sum

def task2F_recursive(arr):
    def even_sum(arr, i=0, sum=0):
        if i < len(arr):
            if arr[i] % 2 == 0:
                sum += arr[i]
            return even_sum(arr, i + 1, sum)
        return sum

    def odd_mul(arr, i=0, mul=1):
        if i < len(arr):
            if arr[i] % 2 != 0:
                mul *= arr[i]
            return odd_mul(arr, i + 1, mul)
        return mul

    return odd_mul(arr) - even_sum(arr)
```

```python
class Node:
  def __init__(self, value=None, next = None):
    self.elem = value
    self.next = next
```

```python
def showLL(head):
  if head!=None:
    n = head
    while n!=None:
        print(n.elem, end=' -> ')
        n = n.next
    print()

def arr2LL(arr):
  head = Node(arr[0])
  n = head
  for i in range(1,len(arr)):
      newNode = Node(arr[i])
      n.next = newNode
      n = newNode
  tail = n
  return head
```

```python
# Task - 03:

def task3A( head ):
    current = head
    while current != None:
        print(current.elem,end="  ")
        current = current.next

def task3B_recursive( head ):
    if head != None:
        print(head.elem,end="  ")
        head = head.next
        task3B_recursive( head )

def task3C( head , sum =0):
    current = head
    while current != None:
        sum += current.elem
        current = current.next
    return sum

def task3D_recursive( head, sum=0):
    if head != None:
        sum += head.elem
        head = head.next
        return task3D_recursive(head, sum)
    return sum

def task3E( head ):
    if head :
        odd_sum = 0
        even_mul = 1
        i = 0
        while head :
            if head.elem % 2 == 0 :
                odd_sum += head.elem
            else :
                even_mul *= head.elem
            head = head.next

        return  even_mul - odd_sum
    return 0

def task3F_recursive(head, odd_sum=0, even_mul=1):
  if not head:
    return even_mul - odd_sum
  if head.elem%2==0:
    odd_sum  += head.elem
  else:
    even_mul *= head.elem
  return task3F_recursive(head.next,odd_sum , even_mul)
```
