```python
import numpy as np
import unittest  # Assuming unittest.output_test() is defined elsewhere

def mergeSortedArray(arr1, arr2):
    size1, size2 = len(arr1), len(arr2)
    merged_size = size1 + size2
    merged_arr = np.array([0] * merged_size)  # Initialize merged array

    i = j = k = 0

    # Merge two sorted arrays into merged_arr
    while i < size1 and j < size2:
        if arr1[i] <= arr2[j]:
            merged_arr[k] = arr1[i]
            i += 1
        else:
            merged_arr[k] = arr2[j]
            j += 1
        k += 1

    # Copy remaining elements from arr1
    while i < size1:
        merged_arr[k] = arr1[i]
        i += 1
        k += 1

    # Copy remaining elements from arr2
    while j < size2:
        merged_arr[k] = arr2[j]
        j += 1
        k += 1

    return merged_arr  # Return the manually merged sorted array

# Driver Code
a1 = np.array([1, 2, 3])
print(f'Sorted Array 1: {a1}')
a2 = np.array([2, 5, 6])
print(f'Sorted Array 2: {a2}')
returned_value = mergeSortedArray(a1, a2)
print(f'Merged Sorted Array: {returned_value}\n')  # Expected: [1, 2, 2, 3, 5, 6]
unittest.output_test(returned_value, np.array([1, 2, 2, 3, 5, 6]))

print('\n==================================\n')

a3 = np.array([1, 3, 5, 11])
print(f'Sorted Array 3: {a3}')
a4 = np.array([2, 7, 8])
print(f'Sorted Array 4: {a4}')
returned_value = mergeSortedArray(a3, a4)
print(f'Merged Sorted Array: {returned_value}\n')  # Expected: [1, 2, 3, 5, 7, 8, 11]
unittest.output_test(returned_value, np.array([1, 2, 3, 5, 7, 8, 11]))
```

```python
#Assignment-02

#1. Check Even or Odd: Write a program to check if a given number is even or odd.

var = input("Enter a number : ")
if var.lstrip('-').isnumeric() or var.isnumeric():
  var = int(var)
  if var%2 == 0:
    print(f"{var} is EVEN!")
  else:
    print(f"{var} is ODD!")
else:
  print(f"{var} is an invalid input!")
```

```python
# 2. Positive, Negative, or Zero: Determine if a number is positive, negative, or zero.

var = input("Enter a number : ")
if var.lstrip('-').isnumeric() or var.isnumeric():
  var = int(var)
  if var < 0:
    print(f"{var} is a Negative number!")
  elif var > 0:
    print(f"{var} is Positive number!")
  else :
    print("The entered number is ZERO!")
else:
  print(f"{var} is an invalid input!")
```

```python
# 3. Vowel or Consonant: Write a program to check whether a character is a vowel or consonant.

char = input("Enter a single character: ").lower()

if len(char) == 1 and char.isalpha():
    if char in "aeiou":
        print(f"{char} is a VOWEL!")
    else:
        print(f"{char} is a CONSONANT!")
else:
    print("Invalid input! Please enter a single alphabet character.")
```

```python
# 4. Leap Year: Check if a given year is a leap year or not.

year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a Leap Year!")
else:
    print(f"{year} is NOT a Leap Year.")
```

```python
#5. Grade Evaluator: Given a score, print the grade (A, B, C, D, or F) based on predefined ranges.

score = input("Enter the score: ")

if score.isnumeric():
  score = float(score)
  if 90 <= score <= 100:
      print("Grade: A")
  elif 80 <= score < 90:
      print("Grade: B")
  elif 70 <= score < 80:
      print("Grade: C")
  elif 60 <= score < 70:
      print("Grade: D")
  else:
      print("Grade: F")
else:
    print("Invalid score! Please enter a number between 0 and 100.")
```

```python
import numpy as np

# Define the Node class
class Node:
    def __init__(self, elem):
        self.elem = elem
        self.next = None  # Pointer to the next node

# Function to create a linked list from an array
def createList(arr):
    if len(arr) == 0:
        return None  # If the array is empty, return None

    head = Node(arr[0])  # Create the first node as head
    current = head

    for value in arr[1:]:
        newNode = Node(value)
        current.next = newNode  # Link the new node
        current = newNode  # Move to the new node

    return head  # Return the head of the linked list

# Function to print the linked list
def printLL(head):
    if head is None:
        print("Linked List is empty")
        return

    current = head
    while current is not None:
        print(current.elem, end=" -> ")
        current = current.next
    print("None")

# Function to get the last node (tail)
def get_tail(head):
    if head is None:
        return None

    current = head
    while current.next is not None:
        current = current.next
    return current  # Return the last node

# Function to append a node at the end of the linked list
def appendLL(head, data):
    newNode = Node(data)

    if head is None:  # If list is empty, return newNode as head
        return newNode

    tail = get_tail(head)  # Find the last node
    tail.next = newNode  # Link the new node at the end

    return head  # Head remains unchanged but returned for safety

# Function to check if an element exists in the linked list
def contains(head, target):
    current_node = head
    while current_node is not None:
        if current_node.elem == target:
            return True
        current_node = current_node.next
    return False  # Element not found

# Create a linked list from an array
arr = np.array([10, 20, 30, 40, 50])
head3 = createList(arr)

# Print the linked list
print("Linked List:")
printLL(head3)

# Searching for elements
target = 40
exists = contains(head3, target)
print(f"Contains {target}: {exists}")

target = 60
exists = contains(head3, target)
print(f"Contains {target}: {exists}")

# Handling an empty linked list case
empty_head = None
print("\nTesting empty linked list:")
printLL(empty_head)

target = 40
# exists = contains(empty_head, target)
# print(f"Contains {target}: {exists}")  # Expected output: False

# # Append elements to an empty linked list
# empty_head = appendLL(empty_head, 100)
# print("\nAfter appending to empty linked list:")
# printLL(empty_head)
```
